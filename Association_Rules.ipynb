####As Association of products varies among the countries, i would perform operations for 1 specific country for understanding

```python
fn_data = online_retail_data[online_retail_data['Country'] == 'France']
fn_data
```

```python
fn_data.isna().sum()
```

```python
del fn_data['CustomerID']
```

```python
sa_data = online_retail_data[online_retail_data['Country'] == 'Saudi Arabia']
sa_pivotted_data = pd.pivot_table(data=sa_data,values='Quantity',index='InvoiceNo',columns='Description',aggfunc='sum').fillna(0)
sa_pivotted_data
```

####Implementation of Pivot Table

```python
fr_pivotted_data = pd.pivot_table(data=fn_data,values='Quantity',index='InvoiceNo',columns='Description',aggfunc='sum').fillna(0)
fr_pivotted_data
```

```python
fr_pivotted_one_hot = fr_pivotted_data.applymap(lambda x: 0 if x == 0 else 1)

fr_pivotted_one_hot
```

###Step 4: Data Mining

```python
frequent_itemsets = apriori(fr_pivotted_one_hot,min_support=0.02,use_colnames= True)
frequent_itemsets
```

```python
fr_best_association = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.02)
fr_best_association.head(20)
```

```python
fr_best_association.to_csv('fr_best_anticedents_consequents.csv')
```
