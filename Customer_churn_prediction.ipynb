## Step 4: Data Preparation

```python
cust_churn_data
```

### Dropping Unwantes features and separation of Input and Output Features

```python
cust_churn_data.drop(labels=["RowNumber","CustomerId","Surname"],axis=1,inplace =True)
```

```python
cust_churn_data["Gender"] = cust_churn_data["Gender"].map(
    {
        'Female': 0,
        'Male': 1
    }
)
```

```python
cust_churn_data = pd.get_dummies(
    cust_churn_data,
    columns =['Geography'],
    drop_first=True
)
```

```python
cust_churn_data = cust_churn_data.astype(
    {
        'Geography_Germany': int,
        'Geography_Spain' : int
    }
)
```

```python
cust_churn_data
```

```python
X = cust_churn_data.drop(labels=["Exited"],axis=1)
X
```

```python
y= cust_churn_data["Exited"]
y
```

## Step 5: Model Building

```python
logistic_model = LogisticRegression()
```

##Step 6: Model Training


```python
logistic_model.fit(X,y)
```

##Step 7: Model Testing

```python
y_pred = logistic_model.predict(X)
```

```python
cust_churn_data["Predicted Output"] = y_pred
```

```python
cust_churn_data
```

## Step 9: Model Evaluation

```python
accuracy_score(y,y_pred)
```
