##Step 4: Data Preparation

```python
weather_data.isna().sum()
```

####Since 'Unnamed:0;' is not a necessary and 'ozone' feature has more null values so,dropping unwanted,duplicates and null entries columns

```python
weather_data.drop(labels=['Unnamed: 0','Ozone','Temp C'],axis=1,inplace=True)
```

```python
weather_data[weather_data['Solar.R'].isna()]
```

```python
weather_data.groupby('Weather')['Solar.R'].mean().round()
```

```python
weather_data['Solar.R'] = weather_data.groupby('Weather')['Solar.R'].transform(lambda x: x.fillna(x.mean().round()))
```

```python
weather_data.isna().sum()
```

####Dropping output feature NaN rows

```python
weather_data.dropna(inplace=True)
```

```python
weather_data.isna().sum()
```

```python
weather_data.dtypes
```

```python
weather_data['Month'].value_counts()
```

```python
weather_data['Month'].replace(to_replace='May',value=5,inplace=True)
```

```python
weather_data.head(30)
```

```python
weather_data['Month'] = weather_data['Month'].astype(int)
```

```python
weather_data.dtypes
```

####Label and One Hot Encoder

```python
weather_data_copy01 = weather_data.copy()
```

```python
weather_data_copy01
```

```python
le_encoder = LabelEncoder()
weather_data_copy01['Weather'] = le_encoder.fit_transform(weather_data_copy01['Weather'])
weather_data_copy01
```

#### One Hot Encoding

```python
weather_data_copy_02 = weather_data
```

```python
weather_data_copy_02 = pd.get_dummies(data=weather_data_copy_02,
               columns=['Weather'],
               dtype=int)
weather_data_copy_02
```

####One Hot Encoding using sklearn

```python
weather_data_copy_03 = weather_data
```

```python
ohe = OneHotEncoder(sparse_output=False,dtype=int)
encoded = ohe.fit_transform(weather_data_copy_03[['Weather']])
```

```python
weather_data_copy_03['Weather'].unique()
```

```python
encoded_weather = pd.DataFrame(
   encoded,
   columns= ohe.get_feature_names_out(['Weather']),
   index = weather_data_copy_03.index
)
```

```python
weather_data_copy_03 = pd.concat([weather_data_copy_03.drop(['Weather'],axis=1),encoded_weather],axis=1)
```
