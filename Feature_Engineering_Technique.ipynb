

##Step 2: Data Collection


###Import Necessary Libraries

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel,RFE
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

###Import Dataset

```python
breast_cancer_data = load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_data_df = pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
breast_cancer_data_df
```

##Step 4: Data Preparation

### Separation of Input and Output

```python
breast_cancer_data_df['target'] = breast_cancer_data.target
```

```python
X = breast_cancer_data_df.drop(columns='target')
X
```

```python
y = breast_cancer_data_df['target']
y
```

##Step 5: Model Building

###Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2,random_state=13)
```

```python
rf_model = RandomForestClassifier()
```

##Step 6:Model Training

```python
rf_model.fit(X_train,y_train)
```

##Step 7: Model Testing

```python
y_pred_train = rf_model.predict(X_train)
y_pred_test = rf_model.predict(X_test)
```

##Step 8: Model Evaluation

```python
print('Accuracy Score:',round(accuracy_score(y_test,y_pred_test),4))
print('Precision Score:',round(precision_score(y_test,y_pred_test),4))
print('Recall Score:',round(recall_score(y_test,y_pred_test),4))
print('Confusion Matrix:\n',confusion_matrix(y_test,y_pred_test))
```

###Technique 1 for Feature selection

```python
def runRandomForestClassifier(X_train,X_test,y_train,y_test):
  rf_model_1 = RandomForestClassifier()
  rf_model_1.fit(X_train,y_train)
  rf_model_1.predict(X_train)
  rf_model_1.predict(X_test)
  print('Accuracy Score:',round(accuracy_score(y_test,y_pred_test),4))
  print('Precision Score:',round(precision_score(y_test,y_pred_test),4))
  print('Recall Score:',round(recall_score(y_test,y_pred_test),4))
  print('Confusion Matrix:\n',confusion_matrix(y_test,y_pred_test))
```

```python
sf_model = SelectFromModel(estimator=RandomForestClassifier(),max_features=10)
sf_model.fit(X_train,y_train)
print(sf_model.get_support())
print(X.columns[sf_model.get_support()])
print(len(X.columns[sf_model.get_support()]))

X_train_sf_model =sf_model.transform(X_train)
X_test_sf_model = sf_model.transform(X_test)
```

```python
%%time
runRandomForestClassifier(X_train,X_test,y_train,y_test)
```

```python
%%time
runRandomForestClassifier(X_train_sf_model,X_test_sf_model,y_train,y_test)
```

### Technique 2: RFE

```python
rfe_model = RFE(estimator=RandomForestClassifier(),n_features_to_select=19)
rfe_model.fit(X_train,y_train)
print(rfe_model.get_support())
print(X.columns[rfe_model.get_support()])
print(len(X.columns[rfe_model.get_support()]))
X_train_rfe_model = rfe_model.transform(X_train)
X_test_rfe_model = rfe_model.transform(X_test)
```
