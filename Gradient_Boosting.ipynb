##Step 2: Data Collection

###Import necessary libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_curve,roc_auc_score
```

###Import Dataset

```python
breast_cancer_data = load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_data_df = pd.DataFrame(breast_cancer_data.data,columns = breast_cancer_data.feature_names)
breast_cancer_data_df
```

##Step 3 Data understanding

###Perform initial investigation

##Step 4: Data Preparation

###Implementation of scaler and separation of inputs and output

```python
breast_cancer_data_df['target'] = breast_cancer_data.target
```

```python
X = breast_cancer_data_df.drop(labels='target',axis=1)
X
```

```python
y = breast_cancer_data_df['target']
y
```

```python
std_scaler = StandardScaler()
scaled_X = std_scaler.fit_transform(X)
```

```python
pd.DataFrame(scaled_X, columns=breast_cancer_data.feature_names)
```

##Step 5: Model Building

###train test split

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_X, y, test_size=0.2,random_state=13,shuffle=True,stratify=y)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

###Model Initialization

```python
gradient_boosting_classifier = GradientBoostingClassifier(n_estimators=75,max_depth=3,learning_rate=0.3)
```

##Step 6: Model Training

```python
gradient_boosting_classifier.fit(X_train,y_train)
```

##Step 7: Model Testing

###Training data prediction

```python
y_pred_train = gradient_boosting_classifier.predict(X_train)
```

###Test data Prediction

```python
y_pred_test = gradient_boosting_classifier.predict(X_test)
```

##Step 8: Model Evaluation

###Training Data Evaluation

```python
print('Confusion_matrix\n',confusion_matrix(y_train,y_pred_train))
print('\n',classification_report(y_train,y_pred_train))
print('accuracy Score:',accuracy_score(y_train,y_pred_train))
```

```python
y_prob = gradient_boosting_classifier.predict_proba(X_train)[:,0]
fpr,tpr,threshold = roc_curve(y_train,y_prob,pos_label=0)
print('auc_score:',roc_auc_score((y_train == 0).astype(int),y_prob))
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color='red')
plt.plot([0,1],[0,1],linestyle='--',color='red')
plt.title('ROC curve for training data')
plt.xlabel('False Postive Rates(FPR)')
plt.ylabel('True Postive Rates(TPR)')
plt.show()
```

###Test Data Evaluation

```python
print('Confusion Matrix\n',confusion_matrix(y_test,y_pred_test))
print('\n',classification_report(y_test,y_pred_test))
print('accuracy score:',accuracy_score(y_test,y_pred_test))
```

```python
y_prob = gradient_boosting_classifier.predict_proba(X_test)[:,0]
fpr,tpr,threshold = roc_curve(y_test,y_prob,pos_label=0)
print('auc_score:',roc_auc_score((y_test == 0).astype(int),y_prob))
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color='red')
plt.plot([0,1],[0,1],linestyle='--',color='red')
plt.title('ROC curve for training data')
plt.xlabel('False Postive Rates(FPR)')
plt.ylabel('True Postive Rates(TPR)')
plt.show()
```
