##To understand the users movie similarity we build correlation.

```python
pvt_table = pd.pivot_table(data=movie_data,values='rating',index='movie',columns='userId',aggfunc='sum').fillna(0)
pvt_table.index = movie_data['movie'].unique()
pvt_table.columns = movie_data['userId'].unique()
pvt_table
```

```python
pvt_table.corr().round(3).head(20)
```

```python
movie_data[(movie_data['userId'] == 8) | (movie_data['userId'] == 16)]
```

###Building UBCF Recommendation

####1. Euclidean method

```python
euclidean_pvt_table = pd.pivot_table(data=movie_data,values='rating',index='userId',columns='movie',aggfunc='sum').fillna(0)
euclidean_pvt_table.index =  movie_data['userId'].unique()
euclidean_pvt_table
```

```python
euclidean_correlation = pd.DataFrame(pairwise_distances(X=euclidean_pvt_table,metric='euclidean'))
euclidean_correlation.index = movie_data['userId'].unique()
euclidean_correlation.columns = movie_data['userId'].unique()
euclidean_correlation.head(20)
```

```python
movie_data[(movie_data['userId'] == 8) | (movie_data['userId'] == 16)]
```

####2.Cosine Correlation

```python
cosine_correlation = pd.DataFrame(1-pairwise_distances(X=euclidean_pvt_table,metric='cosine'))
cosine_correlation.index = movie_data['userId'].unique()
cosine_correlation.columns = movie_data['userId'].unique()
cosine_correlation
```

###observing the distances to find relationship
