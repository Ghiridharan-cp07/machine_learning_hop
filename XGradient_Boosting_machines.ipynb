## Step 2: Data Collection

###Import Necessary Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
import xgboost as xgb
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_curve,roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

###Import Dataset

```python
breast_cancer_data = load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_data_df = pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
```

##Step 4: Data Preparation

####Separation of input and output feature and implemenatation of standardscaler

```python
breast_cancer_data_df['target'] = breast_cancer_data['target']
```

```python
X = breast_cancer_data_df.drop(labels='target',axis=1)
X
```

```python
y = breast_cancer_data_df['target']
y
```

```python
std_scaler = StandardScaler()
scaled_X = std_scaler.fit_transform(X)
pd.DataFrame(scaled_X ,columns = breast_cancer_data.feature_names)
```

##Step 5: Model Building

###Train test split

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_X, y, test_size=0.2,random_state=26,shuffle=True,stratify=y)
```

###Model Initialization

```python
xgboost_classifier = xgb.XGBClassifier(n_estimators=100,max_depth=3,learning_rate=0.2,lamda=5,gamma=1)
```

###GridSearchCV

```python
grid_search = GridSearchCV(estimator=xgboost_classifier,
                           param_grid={
                               'max-depth':[3,4,5],
                               'learning_rate':[0.05,0.1,0.2,0.3],
                               'n_estimators':[100,200],
                               'lambda':[1,1.5,5,10],
                               'gamma':[1,1.5,2,3,4]
                           }
)
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

##Step 6: Model Training

```python
xgboost_classifier.fit(X_train,y_train)
```

##Step 7: Model Testing

###Training data prediction

```python
y_pred_train = xgboost_classifier.predict(X_train)
```

###Test Data Prediction

```python
y_pred_test = xgboost_classifier.predict(X_test)
```

##Step 8: Model Evaluation

###Training Data Evaluation

```python
print('confusion matrix\n',confusion_matrix(y_train,y_pred_train))
print(classification_report(y_train,y_pred_train))
print('accuracy score:',accuracy_score(y_train,y_pred_train))
```

###Test Data Evaluation

```python
print('confusion matrix\n',confusion_matrix(y_test,y_pred_test))
print(classification_report(y_test,y_pred_test))
print('accuracy_score:',accuracy_score(y_test,y_pred_test))
```
