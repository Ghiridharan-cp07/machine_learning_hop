###Import Necessary Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
```

```python
house_data = fetch_california_housing()
house_data
```

```python
house_data_df = pd.DataFrame(house_data.data,columns=house_data.feature_names)
house_data_df
```

```python
X = house_data_df
X = X.head(350)
X.shape
```

```python
X
```

```python
y = house_data.target
y = y[:350]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X ,y, test_size=0.2, random_state=13)
```

###Regularization model Training



```python
linear_model = LinearRegression()
ridge_model = Ridge(alpha=1.0)
lasso_model = Lasso(alpha=0.1)
elastic_model = ElasticNet(alpha=0.1,l1_ratio=0.5)
```

```python
linear_model.fit(X_train,y_train)
ridge_model.fit(X_train,y_train)
lasso_model.fit(X_train,y_train)
elastic_model.fit(X_train,y_train)
```

```python
print('Linear coefficients\n')
print(linear_model.coef_)
print('\n')

print('Rigde coeffecients\n')
print(ridge_model.coef_)
print('\n')

print('Lasso coefficients\n')
print(lasso_model.coef_)
print('\n')

print('ElasticNet coefficient\n')
print(elastic_model.coef_)
```

###Plotting the coefficients

```python
def plot_coefficients(model,feature_names,title):

  coef_df = pd.DataFrame({
      'features': feature_names,
      'coefficients':model.coef_
  })


  coef_df =  coef_df.sort_values(by='coefficients')
  print(coef_df)
  plt.figure(figsize=(10,6))
  plt.barh(coef_df['features'],coef_df['coefficients'])
  plt.title(title)
  plt.xlabel('Coefficients')
  plt.ylabel('Features')
  plt.show()

plot_coefficients(
    linear_model,
    X.columns,
    'Linear Regression Coefficients'
)

plot_coefficients(
    ridge_model,
    X.columns,
    'Ridge model Coefficients'
)

plot_coefficients(
    lasso_model,
    X.columns,
    'Lasso model Coeffiecients'
)

plot_coefficients(
    elastic_model,
    X.columns,
    'elacticnet model Coeffiecients'
)
```

```python
models = {'Linear Model': linear_model,
'Ridge Model': ridge_model,
'Lasso Model': lasso_model,
'Elastic Model':elastic_model}

for name,model in models.items():

  y_pred = model.predict(X_test)

  mse = mean_squared_error(y_test,y_pred)
  r2 = r2_score(y_test,y_pred)
  print(name)
  print('Mean_squared_error:',mse)
  print('R2 Score:',r2)
  print("--------------------")
```
